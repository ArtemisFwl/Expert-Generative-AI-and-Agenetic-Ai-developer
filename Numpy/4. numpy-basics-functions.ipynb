```python
np.round(k1*100)
```

### in numpy, 0= columns and 1= rows

```python
# if we want maximum of every row 

np.max(k1, axis=1)
```

```python
# if we want maximum of every column 
np.max(k1, axis=0)
```

```python
# Product of every column 
np.prod(k1, axis=0)
```
