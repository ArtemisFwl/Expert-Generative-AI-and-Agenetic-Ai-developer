## While loop

```python
i=1

while i<=5:
    print('data science:',i)
    i=i+1
```

```python
i=5

while i>0:
    print('data science:',i)
    i=i-1
```

## Nested While 

```python
i=1 

while i<=5:
    print("Data Science")
    
    j=1
    while j<=4:
        print('Technology')
        j=j+1
        
    i=i+1
    print()
```

```python
i=1 

while i<=5:
    print("Data Science", end="") # when we put end, new line is not created
    
    j=1
    while j<=4:
        print('Technology',end="") # when we put end, new line is not created
        j=j+1
        
    i=i+1
    print()
```

```python
i=1
while i<=2:
    j=0
    
    while j<=2:
        print(i*j, end=" ")
        j+=1
    print()
    i+=1
```

```python
i=1
while i<=4:
    j=0
    while j<=3:
        print(i*j, end= "")
        j+=1
    print()
    
    i+=1
        
```

```python
for i in range (1,21):
    if i%2 ==0:
        
        print(i, end=" ")
```

## Break, Continue and Pass 

```python
# I want only 5 numbers 

for i in range (1,11):
    if i==6:
        break
    print(i, end=" ")
```

```python
# I want only 5 numbers 

for i in range (1,11):
    if i==6:
        continue
    print(i, end=" ")
    
# 6 is not displayed 
```

```python
for i in range (1,11):
    pass

# Used in abstraction in OOPS concept
```
