## Data Cleaning or Data Cleansing

```python
emp['Name']
```

```python
emp["Name"]=emp["Name"].str.replace(r'\W','', regex=True) # \W means non word characters
```

```python
emp["Name"]
```

```python
emp["Domain"]=emp["Domain"].str.replace(r'\W','', regex=True) # \W means non word characters
```

```python
emp["Salary"]=emp["Salary"].str.replace(r'\W','', regex=True) # \W means non word characters
```

```python
emp.head()
```

```python
emp["Age"]=emp["Age"].str.replace(r'\W','', regex=True) # \W means non word characters
```

```python
emp['Age']=emp['Age'].str.extract('(\\d+)')
```

```python
emp['Exp']=emp['Exp'].str.extract('(\\d+)')
```

## EDA

### 1) Missing Vlaue treatment

```python
clean_data['Age']
```

```python
import numpy as np
```

```python
clean_data["Age"]=clean_data["Age"].fillna(np.mean(pd.to_numeric(clean_data["Age"])))
```

```python
clean_data["Age"]
```

```python
clean_data["Exp"]=clean_data["Exp"].fillna(np.mean(pd.to_numeric(clean_data["Exp"])))
```

```python
clean_data["Location"]=clean_data["Location"].fillna((clean_data["Location"].mode()[0]))
```
